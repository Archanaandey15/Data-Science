```python
plt .figure(figsize=(6,5))
sns.histplot(df['math score'],bins=12,kde=True,color='purple')
plt.title('mathmarks dist')
plt.xlabel('frequency')
plt.ylabel('mathmarks')
plt.show()

```

```python
plt.figure(figsize=(12,6))
sns. boxplot(x=df['reading score'],y=df['writing score'],color='blue')
plt.title('boxplot')
plt.xlabel('reading score')
plt.ylabel('writing score')
plt.show()
    
        
    
        
            
```
